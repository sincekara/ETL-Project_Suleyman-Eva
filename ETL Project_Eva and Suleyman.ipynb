```python
#Reading the csv file regarding the Houston schools downloaded from "https://data.world/"
school_data = pd.read_csv("public_schools_in_houston.csv")
school_data.head()
```

```python
#Selecting the necessary columns from the data set.
school = school_data[["ZIP", "DISTNAME", "CAMPNAME", "GRADERANGE"]]
school.head()
```

```python
#Renaming the columns.
school_new = school.rename(columns = {"ZIP":"Zipcodes", "DISTNAME":"School District", "CAMPNAME":"Campus Name", "GRADERANGE":"School Type"})
school_new.head()
```

```python
#Sorting the zipcodes.
school_sorted = school_new.sort_values("Zipcodes", ascending=True)
school_sorted.head()
```

```python
#Extracting the related zipcodes from the Houston schools dataset.
selected_df = school_sorted[school_sorted['Zipcodes'].isin(census_pd_sorted_new.Zipcodes.tolist())]
selected_df.head()
```

```python
#Reseting the index.
new_selected = selected_df.reset_index()
new_selected.head()
```

```python
#Dropping the repeated index column to have a cleaner data set.
new_selected.drop(columns = "index", inplace=True)
```

```python
new_selected.head()
```

```python
#Grouping the data based on zipcodes and countig the schools in each unique zipcode.
clean = new_selected.groupby("Zipcodes").count()
clean.head()
```

```python
#Dropping the unused columns.
clean.drop(columns = ["School District", "Campus Name"], inplace=True)
cleanest = clean.reset_index()
cleanest.head()
```

```python
#Renaming the columns to see how many school are there in each zipcode in Harris County.
cleanest.rename(columns = {"School Type":"Number of Schools"}, inplace=True)
cleanest.head()
```

```python
#Merging the two datasets including the Harris County's population and Harris County's schools on zipcodes.
merged_data = pd.merge(census_pd_sorted_new, cleanest, on= "Zipcodes", how = "right")
merged_data.head()
```

```python
#Finally to calculate the amount of population per school within given zipcodes, 
#a new column added to the merged_data dataset. To get the data for the new column the population is devided by
#the number of schools in each zipcodes.
```

```python
merged_data["Population per School"] = merged_data["2017 Population"] / merged_data["Number of Schools"]
merged_data
```

```python
#Finding the total number of population in Harris County
pop_per_school = merged_data["2017 Population"].sum() / merged_data["Number of Schools"].sum()
print(pop_per_school)
```
